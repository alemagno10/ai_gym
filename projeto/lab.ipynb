```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("lab.xlsx")
df = df.set_index("Objetivo")
df.head(12)
```

```python
plt.figure()
plt.plot(df["Largura"], label="Largura")
plt.plot(df['Profundidade 10'], label="Profundidade 10")
plt.plot(df['Profundidade 100'], label="Profundidade 100")
plt.plot(df['Profundidade Iterativa'], label="Profundidade Iterativa")
plt.legend()
plt.title("Comparação de tempo de execução entre algoritmos de busca")
plt.ylabel("Tempo de Execução (segundos)")
plt.xlabel("Objetivo")
plt.xticks(list(range(0,51,5)))
plt.show()
```

1. Os algoritmos de Largura e Profundidade Iterativa são considerados ótimos, de acordo com o que foi discutido em sala. Isso ocorre pois utilizam um mecanismo de busca horizontalizada, ou seja, buscam o resultado pelas camadas mais superiores.

2. De acordo com o discutido em aula, para uma busca ser considerada completa ela deve encontrar a resposta sempre que houver uma. Esse comportamento foi observado nos tipos: Largura, Profundidade 100 e Iterativa, pois independentemente do tempo ou de memória consumidas, foram capazes de encontrar a solução.

3. A Busca em Largura para um objetivo de 50 travou durante o processo por falta de memória, pois esse tipo de busca apesar de ótimo e completo, necessita guardar toda a árvore de sucessores para contabilizar o caminho. E durante este exemplo, a árvore excedeu o limite de memória da máquina.

4. O Busca em Profundidade 100 foi completo e conseguiu encontrar o resultado num menor intervalo de tempo. Isso ocorreu em função da simplicidade do programa, sendo possível resolvê-lo em uma única árvore, na maioria dos casos.
